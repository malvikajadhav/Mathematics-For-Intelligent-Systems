## I have considered random values for dimentions of the matrices A,b,C,d since no values have been specified in the problem statement 

```julia
m=100
n=250
p=50
A = randn(m,n)
b = randn(m)
C = randn(p,n)
d = randn(p)
```

```julia
# KKT equation for Constrained Least Squares
xzhat = [2*A'*A C'; C zeros(50,50)] \ [2*A'*b; d] #Since given that associated KKT matrix is invertible
xhat = xzhat[1:250,:]
print(xhat)
```
